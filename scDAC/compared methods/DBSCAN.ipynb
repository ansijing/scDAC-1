```python
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import normalized_mutual_info_score
from sklearn import metrics
from sklearn.mixture import BayesianGaussianMixture
from sklearn.metrics import silhouette_score
from modules import models, utils
import torch as th
from numpy import *
from sklearn.cluster import DBSCAN
import numpy as np
```

```python

path_z = '/scDAC-main/scDAC/result/baron_single/e0/default/represent/z.csv'
path_label = './data/b1/label_seurat/l1.csv'
z_list = utils.load_csv(path_z)
z_np = np.array(z_list, dtype=np.float64)
epsarange = arange(0.01, 10, 0.05)
for ceps in epsarange:
    for cmin_samples in arange(1, 100):
        bgm = DBSCAN(eps=ceps, min_samples=cmin_samples).fit(z_np)
        predict_label = bgm.labels_
        label_pre_ten = th.Tensor(predict_label).unsqueeze(1)
        label_true = utils.load_csv(path_label)
        label_tlist = utils.transpose_list(label_true)[1][1:]
        label_plist = utils.transpose_list(label_pre_ten)[0]
        ari = adjusted_rand_score(label_tlist, label_plist) #l1 kpca20
        nmi = normalized_mutual_info_score(label_tlist, label_plist)
        # sc = silhouette_score(z_np, label_plist)
        # sc1 = (sc +1)/2
        print("eps", ceps,"min_samples",cmin_samples,  "ari:", ari, "nmi:", nmi)
        utils.save_tensor_to_csv(label_pre_ten, "./data/baron_single/label_DBSCANae.csv")
```
